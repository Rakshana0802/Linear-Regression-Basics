```python
import pandas as pd
from sklearn import datasets #lib for python and used for all necessary functions of data science
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#predefined and preprocessed dataset loaded from sklearn of boston housing
var = datasets.load_boston() 
```

```python
var
```

```python
x = var.data
```

```python
#in form of 2D matrices coz of multiple X
print(x) 
```

```python
y = var.target
```

```python
#in form of 1D matrices coz only one Y
print (y) 
```

```python
#coverted semistructured data to structured dataframe data format
boston = pd.DataFrame(x, columns=var.feature_names) 
boston.head(10)
```

```python
#column insertion in dataframe
boston['MEDV']=y 
boston.head(10)
```

```python
#checking number of null entries
boston.isnull().sum()
```

```python
#splitting data into x and y (also see next code for y split)
x=boston.drop(columns=['MEDV']) 
```

```python
print (x)
```

```python
y=boston['MEDV']
print (y)
```

```python
#splitting data into training and testing
from sklearn.model_selection import train_test_split 

#test_train_split splits data into 4 parts i.e given in LHS(position of each is mandatory). 
#it takes 3 arguments...test_size 0.2 is the testing part and 0.8 will be the training part
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) 
```

```python
#linear_model is sublib
from sklearn import linear_model 
```

```python
#linear regression function to create linear regression model(lr) which is a naive model
lr=linear_model.LinearRegression() 
```

```python
#train the model-fitting the model
lr.fit(x_train,y_train) #fit is predefined fn of linear_model to convert naive model to trained model
```

(n_jobs is for changing no of cpu core)

```python
#to find prdicted target (yhat) using created model (lr)
y_predict=lr.predict(x_test) 
print (y_predict)
```

```python
print (y_test)
```

(now as shape od y_predict and y_test is nor same... hence reshaping has to be done now to be able to compare the two values) will be done later :3

```python
#prints beta1,beta2,beta3.....beta n
print (lr.coef_) 
```

```python
#print beta0
print (lr.intercept_) 
```

```python
#import fn mean_squared_error from sublib metrices of lib sklearn
from sklearn.metrics import mean_squared_error 
```

```python
mean_squared_error(y_predict,y_test) #fn to find MSE
```

error is high hence approch needs to be changed
